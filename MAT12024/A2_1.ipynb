```python
class Student:
    def __init__(self, sid, name, math, literature, english, history, physics, school):
        '''
        Khởi tạo đối tượng Student với các thuộc tính:
        - sid: mã sinh viên (str) : sid có dạng A012345 trong đó:
            + Ký tự đầu tiên đại diện khối thi
            + 2 ký tự tiếp theo là mã tỉnh của sinh viên
                . mã tỉnh thuộc 1 trong 4 mã : 59 (Tp Hồ Chí Minh), 12 (Lạng Sơn), 99 (Bắc Ninh), 34 (Hải Dương)
                . Trong đó 59 là khu vực 1, 34 99 là khu vực 2, 12 là khu vực 3
        - name: tên sinh viên (str)
        - math: điểm môn Toán (float) [1-10]
        - literature: điểm môn Văn (float) [1-10]
        - english: điểm môn Tiếng Anh (float) [1-10]
        - history: điểm môn Lịch sử (float) [1-10]
        - physics: điểm môn Vật lý (float) [1-10]
        - school: tên trường (str) : trường mà thí sinh đăng kí thi vào , gồm 3 trường [hus, neu, ftu]
        '''
        self.sid = sid
        self.name = name
        self.math = math
        self.literature = literature
        self.english = english
        self.history = history
        self.physics = physics
        self.school = school
        self.__group_score = 0.0
        self.update_score()
        def get_group(self):
            return self.sid[0]
        def get_province(self):
            return int(self.sid[1:3])
        def get_id(self):
            return self.sid
        def get_name(self):
            return self.name
        def get_school(self):
            return self.school
        def set_group_score(self, score):
            return self.__group_score == score
        def get_group_score(self):
            return self.__group_score
        def set_scores(self, math, literature, english, history, physics):
            '''
            Cập nhật điểm các môn học
            '''
            self.math = math
            self.literature = literature
            self.english = english
            self.history = history
            self.physics = physics
            return self
        def get_scores(self):
            '''
            Trả về điểm các môn học
            '''
            return [self.math, self.literature, self.english, self.history, self.physics]
        def get_base_group_score(self):
            '''
            Hàm trả về điểm tuyển sinh cơ bản theo khối của sinh viên
            Khối A: Toán*2 + Lý + Hóa
            Khối B: Văn *2 + sử + anh
            Khối C: sử *2+văn + anh
            Khối D: văn *2 + sử + toán
            '''
            return 0.0
        def update_score(self):
            '''
            hàm tính và cập nhật điểm thi cho thí sinh vào thuộc tính self.__group_score
            điểm thi được cập nhật dựa trên điểm tuyển sinh cơ bản theo khối của sinh viên
            nếu thí sinh có điểm thi thuộc khu vực 1 thì được không được cộng điểm
            nếu thí sinh có điểm thi thuộc khu vực 2 thì được cộng 0.5 điểm
            nếu thí sinh có điểm thi thuộc khu vực 3 thì được cộng 1 điểm
            điểm sau khi được cộng cần được cập nhật bằng hàm set_group_score
            '''
        def get_avg_score(self):
            '''
            Hàm trả về điểm trung bình của sinh viên
            '''
            return sum(self.get_scores()) / len(self.get_scores())
        def __str__(self):
            '''
            Hàm trả về chuỗi mô tả sinh viên
            '''
            return f"Student({self.sid}, {self.name}, {self.get_scores()}, {self.school})"
            
```

```python
from Student import Student
def read_students_from_file(filename):
    '''
    Hàm đọc danh sách sinh viên từ file
    File có định dạng:
    sid,name,math,literature,english,history,physics,school
    '''
    students = []
    with open(filename, 'r') as f:
        for line in f:
            parts = line.strip().split(',')
            if len(parts) != 8:
                continue
            sid, name, math, literature, english, history, physics, school = parts
            student = Student(sid, name, float(math), float(literature), float(english), float(history), float(physics), school)
            students.append(student)
    return students
```
